 ### Dickey-Fuller test
 
 **H_0**: la serie de tiempo no es estacionaria
 **H_1**: la serie de tiempo es estacionaria

```python
result = adfuller(BTC_df.close)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

Dado que el p_valor es > 0.05 no rechazamos nuestra hiptesis inicial con una significancia de 0.05, por tanto, esto se trata de una serie no estacionario

### ARIMA

```python
best_aic = np.inf
best_bic = np.inf

best_order = None
best_mdl = None

pq_rng = range(5)
d_rng  = range(3)

for i in pq_rng:
    for d in d_rng:
        for j in pq_rng:
            try:
                # print(i, d, j)
                tmp_mdl = ARIMA(train, order=(i,d,j)).fit()
                tmp_aic = tmp_mdl.aic
                if tmp_aic < best_aic:
                    best_aic = tmp_aic
                    best_order = (i, d, j)
                    best_mdl = tmp_mdl
            except: continue
print('aic: {:6.5f} | order: {}'.format(best_aic, best_order))
```

 el mejor modelo ARIMA encontrado tiene un AIC de aproximadamente 32.83412 con un orden de (0, 1, 0). Esto sugiere que el mejor modelo es un ARIMA simple que solo necesita diferenciacin de primer orden para hacer la serie estacionaria, y no necesita trminos AR (autoregresivos) ni MA (de media mvil)

#### Auto ARIMA


```python
model = pm.arima.auto_arima(train, start_p=1, start_q=1,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p=3, max_q=3, # maximum p and q
                      m=1,              # frequency of series
                      d=None,           # let model determine 'd'
                      seasonal=False,   # No Seasonality
                      start_P=0, 
                      D=0, 
                      trace=True,
                      error_action='ignore',  
                      suppress_warnings=True, 
                      stepwise=True)
```

El modelo con el valor ms bajo de AIC es ARIMA(0,1,0)(0,0,0)[0], con un AIC de 32.834. Dado que los otros modelos tienen valores ms altos de AIC, el modelo ARIMA(0,1,0) sin trminos estacionales es considerado el mejor modelo entre los evaluados. 

```python
model = ARIMA(train, order=best_order)
model_fit = model.fit()
```

```python
best_aic = np.inf
best_bic = np.inf
best_hqic = np.inf


best_aic_order = None
best_bic_order = None
best_hqic_order = None

pq_rng = range(5) 
d_rng = range(3)   
for i in pq_rng:
    for d in d_rng:
        for j in pq_rng:
            try:
            
                tmp_mdl = ARIMA(train, order=(i,d,j)).fit()
                tmp_aic = tmp_mdl.aic
                tmp_bic = tmp_mdl.bic
                tmp_hqic = tmp_mdl.hqic
                

                if tmp_aic < best_aic:
                    best_aic = tmp_aic
                    best_aic_order = (i, d, j)
                
                
                if tmp_bic < best_bic:
                    best_bic = tmp_bic
                    best_bic_order = (i, d, j)
         
                if tmp_hqic < best_hqic:
                    best_hqic = tmp_hqic
                    best_hqic_order = (i, d, j)
                    
            except: 
                continue


results_df = pd.DataFrame({
    'AIC': [best_aic],
    'BIC': [best_bic],
    'HQIC': [best_hqic],
    'AIC Order': [best_aic_order],
    'BIC Order': [best_bic_order],
    'HQIC Order': [best_hqic_order]
})

print(results_df)
```

AIC (32.83412): Este es el ms bajo de los tres criterios, lo cual es favorable, ya que indica que el modelo tiene un buen equilibrio entre la bondad de ajuste y la simplicidad. En el contexto de las finanzas y los mercados de criptomonedas, que son notoriamente voltiles y ruidosos, un AIC bajo sugiere que el modelo podra generalizar bien en nuevos datos, evitando el sobreajuste.

BIC (36.267842): El BIC es ms alto que el AIC, lo que es tpico porque el BIC penaliza los modelos con ms parmetros ms fuertemente que el AIC. A pesar de ser ms alto, el hecho de que no sea significativamente mayor que el AIC sugiere que la penalizacin adicional por la complejidad del modelo no est sugiriendo que el modelo sea excesivamente complejo. Esto es positivo en el modelado de series de tiempo financieras, ya que los modelos ms sencillos son menos propensos a sobreajustarse a las peculiaridades de los datos histricos.

HQIC (34.219369): El HQIC se sita entre el AIC y el BIC en trminos de valor. Al igual que el BIC, tambin penaliza la complejidad del modelo, pero no tanto como el BIC. Un HQIC relativamente bajo junto con un AIC bajo sugiere nuevamente que el modelo ha encontrado una buena mezcla de complejidad y capacidad de ajuste.

### Rolling

```python
def arima_rolling(history, test):
    
    predictions = list()
    for t in range(len(test)):
        model = ARIMA(history, order=best_order)
        model_fit = model.fit()
        output = model_fit.forecast()
        yhat = output[0]
        predictions.append(yhat)
        obs = test[t]
        history.append(obs)
        print('predicted=%f, expected=%f' % (yhat, obs))
        
    return predictions
test_2wl = test_2w.tolist()
yhat_2w  = arima_rolling(train.tolist(), test_2wl)
```

### Tablas de error

```python
def forecast_accuracy(forecast, actual, str_name):
    
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual)) # MAPE
    mae = np.mean(np.abs(forecast - actual))                 # MAE
    rmse = np.mean((forecast - actual)**2)**.5               # RMSE
    mse = np.mean((forecast - actual)**2)                    # MSE
    r2 = r2_score(forecast, actual)
    
    df_acc = pd.DataFrame({'MAE': [mae],
                           'MSE': [mse],
                           'MAPE': [mape],
                           'RMSE': [rmse],
                           'R2': [r2]},
                          index=[str_name])
    
    return df_acc
forecast_accuracy(np.array(test_2wl), np.array(yhat_2w), "21 das")
```

-  Un MAE de 0.122462 sugiere que, en promedio, el modelo se equivoca por aproximadamente 0.122462 unidades en sus predicciones. 
-  Un MSE de 0.021158 es relativamente bajo, lo que podra indicar que el modelo est realizando predicciones precisas.
-  Un MAPE de 0.001332, o 0.1332%, es bastante bajo, lo que indica que el modelo es muy preciso en trminos porcentuales. 
- El RMSE es la raz cuadrada del MSE. Proporciona una medida de la magnitud del error en las mismas unidades que los datos de prediccin. Un RMSE de 0.145459 no es muy alto, lo cual es bueno.
- Un R2 de 0.77959 significa que aproximadamente el 77.959% de la variabilidad de los datos reales se explica por el modelo.

```python
residuos = model_fit.resid

# Grfico de los residuos
plt.figure(figsize=(10,4))
plt.plot(residuos)
plt.title('Residuals')
plt.axhline(0, linestyle='--', color='red')
plt.show()

# Grfico de densidad de los residuos
plt.figure(figsize=(10,4))
residuos.plot(kind='kde')
plt.title('Density Plot of the Residuals')
plt.show()

# Grfico Q-Q para normalidad
sm.qqplot(residuos, line='s')
plt.show()

# ACF y PACF de los residuos
fig, ax = plt.subplots(1,2, figsize=(15,4))
sm.graphics.tsa.plot_acf(residuos, lags=40, ax=ax[0])
sm.graphics.tsa.plot_pacf(residuos, lags=40, ax=ax[1])
plt.show()

# Prueba de Ljung-Box para autocorrelacin
lb_test = acorr_ljungbox(residuos, lags=[10], return_df=True)
print(lb_test)

# Prueba de Dickey-Fuller para estacionariedad
adf_test = sm.tsa.stattools.adfuller(residuos)
print('Test de Dickey-Fuller Aumentado:')
print('ADF Statistic: %f' % adf_test[0])
print('p-value: %f' % adf_test[1])

# Si tienes el mdulo 'scipy', puedes realizar pruebas de normalidad:
from scipy import stats
print('Test de Shapiro-Wilk para normalidad:')
print(stats.shapiro(residuos))
```

Teniendo en cuanta las graficas y puebas estaditicas, los  residuos no presentan autocorrelacin significativa y son estacionarios, lo cual es positivo y sugiere que el modelo ha capturado adecuadamente las dinmicas de los datos.
Sin embargo, los residuos no parecen distribuirse normalmente.

### Forescast 

```python
model = ARIMA(train, order=(0,1,0)) 
results = model.fit()
predictions = results.forecast(steps=len(test_2w))
print('Real:', test_2w)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(train, label='Datos de entrenamiento')
plt.plot(test_2w, label='Datos de prueba')
plt.plot(test_2w.index, predictions, label='Predicciones', color='red')
plt.legend()
plt.title('ARIMA Predicciones')
plt.show()
```

```python
mae = mean_absolute_error(test_2w, predictions)
mse = mean_squared_error(test_2w, predictions)
mape = mean_absolute_percentage_error(test_2w, predictions)
rmse = np.sqrt(mse)
r2 = r2_score(test_2w, predictions)

# Imprimir mtricas de evaluacin
print(f'MAE (Error absoluto medio): {mae}')
print(f'MSE (Error cuadrtico medio): {mse}')
print(f'MAPE (Error absoluto porcentual medio): {mape}')
print(f'RMSE (Raz del error cuadrtico medio): {rmse}')
print(f'R2 (Coeficiente de determinacin): {r2}')

# Obtener informacin sobre el modelo
print(f'BIC (Criterio de Informacin Bayesiano): {results.bic}')
print(f'AIC (Criterio de Informacin de Akaike): {results.aic}')
print(f'HQIC (Criterio de Informacin Hannan-Quinn): {results.hqic}')
```

```python
residuals = results.resid

# Grfico de los residuos
plt.figure(figsize=(10, 6))
plt.plot(residuals)
plt.title('Grfico de Residuos')
plt.show()

# Grfico de densidad de los residuos
plt.figure(figsize=(10, 6))
residuals.plot(kind='kde')
plt.title('Grfico de Densidad de Residuos')
plt.show()

# Grfico Q-Q para normalidad
plt.figure(figsize=(10, 6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title('Grfico Q-Q para Normalidad')
plt.show()

# ACF y PACF de los residuos
plt.figure(figsize=(12, 6))
plot_acf(residuals, ax=plt.subplot(121))
plot_pacf(residuals, ax=plt.subplot(122))
plt.suptitle('ACF y PACF de Residuos')
plt.show()

# Prueba de Ljung-Box para autocorrelacin
lb_test_stat, lb_p_value = acorr_ljungbox(residuals, lags=[10])
print(f'Estadstico de prueba Ljung-Box: {lb_test_stat}')
print(f'p-value: {lb_p_value}')

# Prueba de Dickey-Fuller para estacionariedad
adf_test_stat, adf_p_value, _, _, adf_critical_values, _ = adfuller(residuals)
print(f'Estadstico de prueba Dickey-Fuller: {adf_test_stat}')
print(f'p-value: {adf_p_value}')
print('Valores crticos:')
for key, value in adf_critical_values.items():
    print(f'{key}: {value}')

# Prueba de normalidad
_, normality_p_value = stats.normaltest(residuals)
print(f'p-value de la prueba de normalidad: {normality_p_value}')
```

en este caso, los residuos de nuevo,  no muestran autocorrelacin significativa y son estacionarios, y se observa que los residuos no siguen una distribucin normal como lo compruba el valor del p valor en la prueba de normalidad
